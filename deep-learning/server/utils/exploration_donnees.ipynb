```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configuration pour de meilleurs graphiques
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 6)

# Lire le fichier CSV
df = pd.read_csv('../data/initial-budget-participatif.csv', delimiter=';', encoding='utf-8')

print("Dataset chargé avec succès !")
print(f"Dimensions: {df.shape}")
```

```python
# Colonnes d'intérêt
colonnes_interet = [
    "Titre de l'opération",
    "Edition",
    "Thématique",
    "Direction pilote du projet lauréat",
    "Direction pilote de l'opération",
    "Arrondissement de l'opération",
    "Opération en Quartier Populaire",
    "Budget global du projet lauréat",
    "Avancement de l'opération"
]

# Informations générales
print("=== INFORMATIONS GÉNÉRALES ===\n")
print(f"Nombre total de lignes: {len(df)}")
print(f"Nombre total de colonnes: {len(df.columns)}")
print(f"\nNombre de colonnes d'intérêt: {len(colonnes_interet)}")

# Valeurs manquantes pour les colonnes d'intérêt
print("\n=== VALEURS MANQUANTES ===\n")
for col in colonnes_interet:
    if col in df.columns:
        nb_na = df[col].isna().sum()
        pct_na = (nb_na / len(df)) * 100
        print(f"{col}: {nb_na} ({pct_na:.2f}%)")
    else:
        print(f"{col}: COLONNE INTROUVABLE")

print("\n=== APERÇU DES DONNÉES ===")
df[colonnes_interet].head(10)
```

## Analyse de la colonne 'Thématique'

Cette section analyse la répartition des données dans la colonne 'Thématique' pour identifier les différentes catégories de projets.

```python
# Analyse de la colonne 'Thématique'
print("=== Répartition par Thématique ===\n")

# Compter les occurrences de chaque thématique
thematique_counts = df['Thématique'].value_counts()

print("Liste des thématiques et leur nombre d'occurrences:")
print(thematique_counts)
print(f"\nNombre total de thématiques uniques: {len(thematique_counts)}")

# Créer l'histogramme
plt.figure(figsize=(14, 8))
thematique_counts.plot(kind='bar', color='steelblue', edgecolor='black')
plt.title('Répartition des projets par Thématique', fontsize=16, fontweight='bold')
plt.xlabel('Thématique', fontsize=12)
plt.ylabel('Nombre de projets', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.grid(axis='y', alpha=0.3)
plt.show()

# Afficher les pourcentages
print("\n=== Pourcentages par thématique ===")
thematique_percentages = (thematique_counts / len(df) * 100).round(2)
for thematique, percentage in thematique_percentages.items():
    print(f"{thematique}: {percentage}% ({thematique_counts[thematique]} projets)")
```

```python
# 1. Nombre de lignes dans le fichier
print("=== 1. NOMBRE DE LIGNES ===")
nb_lignes = len(df)
print(f"Nombre total de projets dans le dataset: {nb_lignes:,}")
col_id = "Identifiant de l'opération"
print(f"Nombre de projets uniques: {df[col_id].nunique() if col_id in df.columns else 'N/A'}")
```

```python
# 2. Edition: min et max
print("=== 2. ÉDITION (MIN/MAX) ===")
edition_min = df['Edition'].min()
edition_max = df['Edition'].max()
print(f"Année minimale: {edition_min}")
print(f"Année maximale: {edition_max}")
print(f"Période couverte: {edition_max - edition_min} ans")
print(f"\nNombre d'éditions: {df['Edition'].nunique()}")
print(f"Éditions présentes: {sorted(df['Edition'].unique())}")
```

```python
# 3. Edition: Histogramme du nombre de projets par année
print("=== 3. HISTOGRAMME - PROJETS PAR ANNÉE ===")

projets_par_annee = df['Edition'].value_counts().sort_index()

plt.figure(figsize=(14, 6))
plt.bar(projets_par_annee.index, projets_par_annee.values, color='#3498db', edgecolor='black')
plt.xlabel('Année', fontsize=12, fontweight='bold')
plt.ylabel('Nombre de projets', fontsize=12, fontweight='bold')
plt.title('Nombre de projets du Budget Participatif par année', fontsize=14, fontweight='bold')
plt.xticks(projets_par_annee.index, rotation=45)
plt.grid(axis='y', alpha=0.3)

# Ajouter les valeurs au-dessus des barres
for i, v in enumerate(projets_par_annee.values):
    plt.text(projets_par_annee.index[i], v + 20, str(v), ha='center', fontweight='bold')

plt.tight_layout()
plt.show()

print(f"\n📊 Analyse:")
print(f"- Année avec le plus de projets: {projets_par_annee.idxmax()} ({projets_par_annee.max()} projets)")
print(f"- Année avec le moins de projets: {projets_par_annee.idxmin()} ({projets_par_annee.min()} projets)")
print(f"- Moyenne de projets par an: {projets_par_annee.mean():.0f}")
```

```python
# 4. Direction pilote de l'opération: Pie chart
print("=== 4. DIRECTION PILOTE DE L'OPÉRATION ===")

col_direction = "Direction pilote de l'opération"
direction_operation = df[col_direction].value_counts()

# Afficher le tableau complet de toutes les directions
print("\n📋 Tableau complet de toutes les directions:")
print(f"\n{'Direction':<60} {'Nombre de projets':>20}")
print("=" * 82)
for direction, count in direction_operation.items():
    pct = (count / len(df)) * 100
    print(f"{direction:<60} {count:>10} ({pct:>5.1f}%)")
print("=" * 82)
print(f"{'TOTAL':<60} {direction_operation.sum():>10} (100.0%)")
print(f"\nNombre total de directions différentes: {len(direction_operation)}")

# Préparer les données pour le pie chart: Top 10 + Autres
top_10 = direction_operation.head(10)
autres_count = direction_operation[10:].sum()

# Créer un nouveau Series avec Top 10 + Autres
if autres_count > 0:
    direction_pour_pie = pd.concat([top_10, pd.Series({'Autres': autres_count})])
else:
    direction_pour_pie = top_10

# Pie chart avec Top 10 + Autres
plt.figure(figsize=(12, 8))
colors = sns.color_palette('Set3', len(direction_pour_pie))
plt.pie(direction_pour_pie.values, labels=direction_pour_pie.index, autopct='%1.1f%%', 
        colors=colors, startangle=90, textprops={'fontsize': 9})
plt.title('Répartition des projets par Direction pilote de l\'opération (Top 10 + Autres)', 
          fontsize=14, fontweight='bold')
plt.axis('equal')
plt.tight_layout()
plt.show()

print(f"\n📊 Top 10 des directions:")
for i, (direction, count) in enumerate(top_10.items(), 1):
    pct = (count / len(df)) * 100
    print(f"{i}. {direction}: {count} projets ({pct:.1f}%)")
if autres_count > 0:
    pct_autres = (autres_count / len(df)) * 100
    print(f"\nAutres ({len(direction_operation) - 10} directions): {autres_count} projets ({pct_autres:.1f}%)")
```

```python
# 6. Opération en Quartier Populaire: Pie chart
print("=== 6. OPÉRATION EN QUARTIER POPULAIRE ===")

col_quartier = "Opération en Quartier Populaire"
quartier_pop = df[col_quartier].value_counts()

plt.figure(figsize=(8, 8))
colors = ['#e74c3c', '#3498db', '#95a5a6']
# Créer explode avec la bonne longueur
explode = [0.05] + [0] * (len(quartier_pop) - 1)  # Mettre en évidence la première tranche
plt.pie(quartier_pop.values, labels=quartier_pop.index, autopct='%1.1f%%', 
        colors=colors[:len(quartier_pop)], explode=explode, startangle=90, 
        textprops={'fontsize': 12, 'fontweight': 'bold'})
plt.title('Répartition des projets en Quartier Populaire', fontsize=14, fontweight='bold')
plt.axis('equal')
plt.tight_layout()
plt.show()

print(f"\n📊 Détails:")
for statut, count in quartier_pop.items():
    pct = (count / len(df)) * 100
    print(f"- {statut}: {count} projets ({pct:.1f}%)")
```

```python
# 7. Budget global du projet lauréat: Statistiques et répartition
print("=== 7. BUDGET GLOBAL DU PROJET LAURÉAT ===")

# Nettoyer les données de budget (enlever les NaN)
budget_data = df['Budget global du projet lauréat'].dropna()

# Statistiques
mediane = budget_data.median()
moyenne = budget_data.mean()
minimum = budget_data.min()
maximum = budget_data.max()

print(f"\n💰 Statistiques du budget:")
print(f"- Médiane: {mediane:,.0f} €")
print(f"- Moyenne: {moyenne:,.0f} €")
print(f"- Minimum: {minimum:,.0f} €")
print(f"- Maximum: {maximum:,.0f} €")
print(f"- Écart-type: {budget_data.std():,.0f} €")
print(f"\n- Nombre de projets avec budget: {len(budget_data)}")
print(f"- Nombre de projets sans budget: {df['Budget global du projet lauréat'].isna().sum()}")
```

```python
# 7bis. Budget: Répartition par tranches de prix
print("=== RÉPARTITION DES BUDGETS PAR TRANCHES ===")

# Définir les tranches de budget
tranches = [0, 10000, 50000, 100000, 500000, 1000000, float('inf')]
labels = ['0-10K', '10K-50K', '50K-100K', '100K-500K', '500K-1M', '1M+']

budget_data_copy = budget_data.copy()
budget_tranches = pd.cut(budget_data_copy, bins=tranches, labels=labels, include_lowest=True)
repartition_tranches = budget_tranches.value_counts().sort_index()

# Graphique en barres
plt.figure(figsize=(12, 6))
colors = sns.color_palette('viridis', len(repartition_tranches))
bars = plt.bar(repartition_tranches.index, repartition_tranches.values, color=colors, edgecolor='black')
plt.xlabel('Tranches de budget (€)', fontsize=12, fontweight='bold')
plt.ylabel('Nombre de projets', fontsize=12, fontweight='bold')
plt.title('Répartition des projets par tranches de budget', fontsize=14, fontweight='bold')
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)

# Ajouter les valeurs et pourcentages
for i, (bar, count) in enumerate(zip(bars, repartition_tranches.values)):
    pct = (count / len(budget_data)) * 100
    plt.text(bar.get_x() + bar.get_width()/2, count + 20, 
             f'{count}\n({pct:.1f}%)', ha='center', fontweight='bold')

plt.tight_layout()
plt.show()

print("\n📊 Détails par tranche:")
for tranche, count in repartition_tranches.items():
    pct = (count / len(budget_data)) * 100
    print(f"- {tranche}: {count} projets ({pct:.1f}%)")
```

```python
# 8. Avancement de l'opération: Pie chart
print("=== 8. AVANCEMENT DE L'OPÉRATION ===")

avancement = df['Avancement de l\'opération'].value_counts()

plt.figure(figsize=(10, 8))
colors = sns.color_palette('coolwarm', len(avancement))
plt.pie(avancement.values, labels=avancement.index, autopct='%1.1f%%', 
        colors=colors, startangle=90, textprops={'fontsize': 11, 'fontweight': 'bold'})
plt.title('Répartition des projets par état d\'avancement', fontsize=14, fontweight='bold')
plt.axis('equal')
plt.tight_layout()
plt.show()

print(f"\n📊 Statut des projets:")
for statut, count in avancement.items():
    pct = (count / len(df)) * 100
    print(f"- {statut}: {count} projets ({pct:.1f}%)")

# Analyse
print(f"\n🔍 Analyse:")
if 'FIN' in avancement.index:
    print(f"- {avancement['FIN']} projets terminés sur {len(df)} ({(avancement['FIN']/len(df)*100):.1f}%)")
if 'EN COURS' in avancement.index:
    print(f"- {avancement['EN COURS']} projets en cours ({(avancement['EN COURS']/len(df)*100):.1f}%)")
```

## Synthèse de l'analyse exploratoire

Cette analyse nous a permis de découvrir :

1. **Volume de données** : Le dataset contient plusieurs milliers de projets du Budget Participatif de Paris
2. **Période couverte** : Les projets s'étendent sur plusieurs années d'éditions
3. **Répartition organisationnelle** : Plusieurs directions pilotent les projets (voirie, affaires culturelles, espaces verts, etc.)
4. **Budgets** : Large éventail de budgets avec des projets allant de quelques milliers à plusieurs millions d'euros
5. **Quartiers populaires** : Une partie significative des projets concerne les quartiers populaires
6. **État d'avancement** : Les projets sont à différents stades (terminés, en cours, etc.)

Ces données statistiques peuvent être utilisées pour enrichir les réponses de l'API de prédiction.
